```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Version 1 Cleaned_GenZ_DatingApp_Data.csv")
```

```python
age_bins = [18, 22, 26, 30]
age_labels = ['18-22', '23-26', '27-30']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
time = df.groupby(['Age_Group', 'Primary_App'], observed=True).size().unstack().fillna(0)
```

```python
time.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Dating App Preferences by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Users')
plt.legend(title='Primary App')
plt.show()
```

```python
time = df.groupby(['Age_Group', 'Primary_App'], observed=True).size().unstack().fillna(0)

plt.figure(figsize=(10, 6))
for app in time.columns:
    plt.plot(time.index, time[app], marker='o', label=app)
plt.title('Dating App Trends Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Users')
plt.legend(title='Primary App')
plt.show()
```

```python
# Line Chart: Tracking Daily Usage Time Trends by Age
plt.figure(figsize=(10, 5))
sns.lineplot(x="Age", y="Daily_Usage_Time", data=df, estimator="mean", marker="o", ci=None)
plt.xlabel("Age")
plt.ylabel("Average Daily Usage Time (hours)")
plt.title("Daily Usage Time Trends by Age")
plt.show()
```

```python
# Count of primary app usage per location (city)
location_counts = df["Location"].value_counts().reset_index()
location_counts.columns = ["City", "User_Count"]

# geospatial visualization using bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="User_Count", y="City", data=location_counts, palette="coolwarm")
plt.xlabel("Number of Users")
plt.ylabel("City")
plt.title("Dating App Users Across Different Cities")
plt.show()
```

```python
# Count of primary app usage per age group
app_age_counts = df.groupby(["Age", "Primary_App"]).size().reset_index(name="User_Count")

# Bubble Chart: Representing app popularity across age groups
plt.figure(figsize=(12, 6))
sns.scatterplot(x="Age", y="Primary_App", size="User_Count", data=app_age_counts, legend=False, alpha=0.6, sizes=(20, 500))
plt.xlabel("Age")
plt.ylabel("Primary App")
plt.title("Dating App Popularity Across Age Groups")
plt.show()
```

1, Are younger or older Gen-Z users more active on dating apps?
Younger Gen-Z users (18-22) tend to spend more time daily on dating apps, as seen in the line chart. Daily usage time generally decreases with age.
2, Does dating app preference change over time?
Yes, preferences shift with age. Younger users (18-22) favor Tinder & Bumble for casual dating. Older Gen-Z (23-25) prefer Hinge & OkCupid, which focus more on relationships.
3, Do metro city users behave differently from users in smaller towns?
Yes, metro city users are more active. Higher user count in cities like Delhi, Bangalore, and Kolkata.
