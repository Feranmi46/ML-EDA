```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Version 1 Cleaned_GenZ_DatingApp_Data.csv")
```

```python
# Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=df["Gender"], edgecolor="black", legend = False)
plt.title("Gender Distribution in the Dataset")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=10, kde=True, color="purple", edgecolor="black")
plt.title("Age Distribution of Users")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Location Distribution
plt.figure(figsize=(8, 5))
sns.countplot(y=df["Location"], palette="viridis", edgecolor="black")
plt.title("Location Distribution of Users")
plt.xlabel("Count")
plt.ylabel("Location")
plt.show()
```

```python
# Density Plot: Daily usage time across locations
plt.figure(figsize=(10, 6))
sns.kdeplot(df, x='Daily_Usage_Time', hue='Location', fill=True)
plt.title('Daily App Usage Time Across Locations')
plt.xlabel('Daily Usage Time (Hours)')
plt.ylabel('Density')
plt.show()
```

Discuss ethical considerations in data collection.
When collecting and analyzing data on dating app usage among Gen Z in India, several ethical considerations must be addressed to ensure responsible research practices. Here are key ethical concerns:

1. Privacy & Informed Consent
Users must be aware that their data is being collected, how it will be used, and who will have access to it.
If the dataset includes personal or sensitive information (e.g., gender identity, sexual orientation, or location), explicit consent should be obtained.
Anonymization techniques should be applied to prevent the identification of individuals.
2. Bias & Representation
The dataset should represent a diverse sample of individuals across different demographics (e.g., gender, sexual orientation, region, socioeconomic background).
Underrepresentation of certain groups (e.g., LGBTQ+ users, people from rural areas) can lead to biased conclusions.
3. Stereotyping & Discrimination Risks
The interpretation of data should avoid reinforcing stereotypes (e.g., assumptions about dating behaviors based on gender or region).
Researchers must be cautious about drawing conclusions that could be used to discriminate against certain groups (e.g., assuming certain communities are less interested in online dating).
4. Data Security & Confidentiality
Data should be stored securely to prevent breaches or unauthorized access.
Any personally identifiable information (PII) should be removed or masked to ensure user privacy.
5. Transparency & Purpose of Data Use
The purpose of collecting the data should be clearly defined (e.g., improving dating app features, understanding user behavior).
Data should not be used for unintended purposes, such as targeted advertising or surveillance, without user consent.
